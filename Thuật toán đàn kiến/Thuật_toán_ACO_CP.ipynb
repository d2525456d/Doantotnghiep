```python
# Định nghĩa độ thành thạo của từng công nhân
proficiency_levels = {
    "Lê Thị Phương Thảo": 1.48,
    "Hoàng Thị Mỹ Yến": 1.16,
    "Trần Thị Tuyền": 1.48,
    "Ngô Khắc Hiển": 1.00,
    "Hồ Thị Diệp": 1.16,
    "Nguyễn Thị Mỹ Nhung": 1.16,
    "Thái Thị Nga": 1.16,
    "Trần Thị Cúc": 1.00,
    "Lê Thị Thanh Hà": 1.16,
    "Nguyễn Thị Thúy": 1.16,
    "Trần Thị Cẩm Rin": 1.32,
    "Hoàng Thị Nở": 1.16,
    "Lê Thị Như Quỳnh": 1.16,
    "Trần Thị Thương": 1.16,
    "Đinh Thị Mi": 1.16,
    "Nguyễn Thị Ngọc Lan": 1.16,
    "Bùi Thị Hồng Vân": 1.00,
    "Trần Thị Oanh": 1.16,
    "Nguyễn Thị Gấm": 1.32,
    "Phan Thị Thu Hoài": 1.16,
    "Nguyễn Thị Thu": 1.16,
    "Nguyễn Thị Hài": 1.16,
    "Nguyễn Thị Hồng": 1.48
}
```

```python
# Cấu hình thuật toán đàn kiến
num_ants = 285   # Số lượng kiến
num_iterations = 265  # Số vòng lặp của thuật toán
pheromone_evaporation = 0.2  # Tỷ lệ bay hơi pheromone
pheromone_intensity = 0.7      # Mức độ tăng pheromone
pheromone = np.ones((len(dt), num_workers))   # Khởi tạo ma trận pheromone (Công đoạn x Công nhân)

# Hàm tính tổng lương
def calculate_total_salary(individual, proficiency_levels, dt):
    worker_salaries = {worker: 0 for worker in workers}
    for task_idx, worker_list in enumerate(individual):
        task_price = dt['ĐƠN GIÁ'][task_idx]
        num_workers = len(worker_list)
        price_per_worker = task_price / num_workers
        for worker in worker_list:
            proficiency = proficiency_levels[worker]
            salary_for_task = price_per_worker * proficiency
            worker_salaries[worker] += salary_for_task
    total_salary = sum(worker_salaries.values())
    return total_salary

# Hàm khởi tạo kiến
def initialize_ants(num_ants, pheromone, workers, dt):
    ant_solutions = []
    for _ in range(num_ants):
        solution = []
        worker_machine_assignment = {worker: set() for worker in workers}
        for task_idx in range(len(dt)):
            machine_type = dt['THIẾT BỊ'][task_idx]
            available_workers = []
            for worker in workers:
                machine_types_assigned = worker_machine_assignment[worker]
                if machine_type == 0:
                    if len(machine_types_assigned) < 3 or machine_type in machine_types_assigned:
                        available_workers.append(worker)
                else:
                    if len(machine_types_assigned) < 2 or machine_type in machine_types_assigned:
                        available_workers.append(worker)
            if available_workers:
                probabilities = [pheromone[task_idx][workers.index(worker)] for worker in available_workers]
                selected_workers = random.choices(available_workers, probabilities, k=random.randint(1, 3))
                for worker in selected_workers:
                    worker_machine_assignment[worker].add(machine_type)
                solution.append(list(set(selected_workers)))  # Đảm bảo không trùng công nhân
            else:
                solution.append([])
        ant_solutions.append(solution)
    return ant_solutions

# Hàm kiểm tra ràng buộc và tính toán lương
def fitness_with_machine_limit_aco(individual, proficiency_levels, dt):
    for task_idx, worker_list in enumerate(individual):
        if len(worker_list) == 0 or len(worker_list) > 3:
            return None
    worker_machine_types = {worker: set() for worker in workers}
    for task_idx, worker_list in enumerate(individual):
        machine_type = dt['THIẾT BỊ'][task_idx]
        for worker in worker_list:
            worker_machine_types[worker].add(machine_type)
    penalty = 0
    for worker, machine_types in worker_machine_types.items():
        if 0 in machine_types and len(machine_types) > 3:
            penalty += 1
        elif 0 not in machine_types and len(machine_types) > 2:
            penalty += 1
    if penalty > 0:
        return None
    return calculate_total_salary(individual, proficiency_levels, dt)

# Hàm cập nhật pheromone
def update_pheromone(pheromone, ant_solutions, best_solution, pheromone_evaporation, pheromone_intensity):
    pheromone *= (1 - pheromone_evaporation)
    for task_idx in range(len(dt)):
        for worker_idx, worker in enumerate(workers):
            if worker in best_solution[task_idx]:
                pheromone[task_idx][worker_idx] += pheromone_intensity / calculate_total_salary(best_solution, proficiency_levels, dt)

# Vòng lặp chính của ACO
best_population = []
min_best_wages = float('inf')
for iteration in range(num_iterations):
    ant_solutions = initialize_ants(num_ants, pheromone, workers, dt)
    ant_salaries = []
    for solution in ant_solutions:
        fitness = fitness_with_machine_limit_aco(solution, proficiency_levels, dt)
        if fitness is not None:
            ant_salaries.append((solution, fitness))
    if ant_salaries:
        current_best_solution, current_best_salary = min(ant_salaries, key=lambda x: x[1])
        if current_best_salary < min_best_wages:
            min_best_wages = current_best_salary
            best_population = ant_solutions
            best_solution = current_best_solution
    if best_solution is not None:
        update_pheromone(pheromone, ant_solutions, best_solution, pheromone_evaporation, pheromone_intensity)
    print(f"Thế hệ {iteration + 1}: Lương thấp nhất = {min_best_wages:.4f}")
    if ant_salaries:
        print(f"Thế hệ {iteration + 1}: Lương hiện tại = {current_best_salary:.4f}")

# Lấy giải pháp tốt nhất
best_individual = min(best_population, key=lambda ind: fitness_with_machine_limit_aco(ind, proficiency_levels, dt) or float('inf'))
```

```python
# Phần 6: Hiển thị kết quả chi tiết cho mỗi công nhân

def display_task_allocation_with_time(individual):
    detailed_task_allocation = {worker: [] for worker in workers}
    for task_idx, worker_list in enumerate(individual):
        task_time = dt['TGLT'].iloc[task_idx]
        n_workers = len(worker_list)
        theoretical_time_per_worker = task_time / n_workers
        for worker in worker_list:
            proficiency = proficiency_levels[worker]
            worker_time = theoretical_time_per_worker / proficiency
            detailed_task_allocation[worker].append(f"Công đoạn {dt['TÊN CÔNG ĐOẠN'].iloc[task_idx]}: {worker_time:.2f} giây")
    return detailed_task_allocation

task_allocation_with_time = display_task_allocation_with_time(best_individual)

worker_task_summary = []
for worker, tasks in task_allocation_with_time.items():
    total_time = sum(float(task.split(": ")[1].split()[0]) for task in tasks)
    task_details = " + ".join(tasks)
    worker_task_summary.append({
        'Công Nhân': worker,
        'Chi Tiết Công Việc': task_details,
        'Tổng Thời Gian': total_time
    })

# Chuyển thành DataFrame để hiển thị
worker_task_summary_df = pd.DataFrame(worker_task_summary)

# Hiển thị kết quả chi tiết cho mỗi công nhân
worker_task_summary_df
```
