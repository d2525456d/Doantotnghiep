```python
workers = [
    "Lê Thị Phương Thảo", "Hoàng Thị Mỹ Yến", "Trần Thị Tuyền", "Ngô Khắc Hiển",
    "Hồ Thị Diệp", "Nguyễn Thị Mỹ Nhung", "Thái Thị Nga",
    "Trần Thị Cúc", "Lê Thị Thanh Hà", "Nguyễn Thị Thúy", "Trần Thị Cẩm Rin",
    "Hoàng Thị Nở", "Lê Thị Như Quỳnh", "Trần Thị Thương", "Đinh Thị Mi",
    "Nguyễn Thị Ngọc Lan", "Bùi Thị Hồng Vân", "Trần Thị Oanh",
    "Nguyễn Thị Gấm", "Phan Thị Thu Hoài", "Nguyễn Thị Thu", "Nguyễn Thị Hài",
    "Nguyễn Thị Hồng"
]
num_workers = len(workers)
```

```python
# Định nghĩa độ thành thạo của từng công nhân
proficiency_levels = {
    "Lê Thị Phương Thảo": 1.48,
    "Hoàng Thị Mỹ Yến": 1.16,
    "Trần Thị Tuyền": 1.48,
    "Ngô Khắc Hiển": 1.00,
    "Hồ Thị Diệp": 1.16,
    "Nguyễn Thị Mỹ Nhung": 1.16,
    "Thái Thị Nga": 1.16,
    "Trần Thị Cúc": 1.00,
    "Lê Thị Thanh Hà": 1.16,
    "Nguyễn Thị Thúy": 1.16,
    "Trần Thị Cẩm Rin": 1.32,
    "Hoàng Thị Nở": 1.16,
    "Lê Thị Như Quỳnh": 1.16,
    "Trần Thị Thương": 1.16,
    "Đinh Thị Mi": 1.16,
    "Nguyễn Thị Ngọc Lan": 1.16,
    "Bùi Thị Hồng Vân": 1.00,
    "Trần Thị Oanh": 1.16,
    "Nguyễn Thị Gấm": 1.32,
    "Phan Thị Thu Hoài": 1.16,
    "Nguyễn Thị Thu": 1.16,
    "Nguyễn Thị Hài": 1.16,
    "Nguyễn Thị Hồng": 1.48
}
```

```python
# Thay thế cột "TÊN CÔNG ĐOẠN" bằng chuỗi kết hợp "TÊN CÔNG ĐOẠN" và "Loại Máy"
dt['TÊN CÔNG ĐOẠN'] = dt['TÊN CÔNG ĐOẠN'] + ' - ' + dt['THIẾT BỊ'].astype(str)

# Hiển thị DataFrame kết quả
dt
```

```python
# Các tham số cho thuật toán ACO
pheromone_init = 0.7
pheromone_matrix = np.full((len(tasks_times), len(workers)), pheromone_init)
num_ants = 275             # Tăng số lượng kiến để có thêm các giải pháp trong mỗi thế hệ
num_iterations = 300      # Tăng số vòng lặp để cải thiện khả năng tìm kiếm
evaporation_rate = 0.2      # Giảm tỷ lệ bay hơi để giữ pheromone lâu hơn cho các giải pháp tốt
alpha = 2                 # Giảm nhẹ mức ảnh hưởng của pheromone để tăng sự đa dạng
beta = 12                   # Tăng ảnh hưởng của khoảng cách để chọn các giải pháp ngắn hơn

# Khởi tạo biến toàn cục `worker_machine_assignment`, `worker_task_count`, và `worker_assigned`
worker_machine_assignment = {worker: set() for worker in workers}
worker_task_count = {worker: 0 for worker in workers}
worker_assigned = {worker: 0 for worker in workers}

# Hàm tính thời gian từng công đoạn dựa trên phân bổ công nhân
def calculate_stage_times(individual):
    stage_times_summary = []
    for task_idx, workers_in_task in enumerate(individual):
        task_time = dt['TGLT'].iloc[task_idx]
        n_workers = len(workers_in_task)
        if n_workers > 0:
            theoretical_time_per_worker = task_time / n_workers
            total_time = sum(theoretical_time_per_worker / proficiency_levels[worker] for worker in workers_in_task)
        else:
            total_time = 0
        stage_times_summary.append({"Công Đoạn": dt['TÊN CÔNG ĐOẠN'][task_idx], "Tổng Thời Gian": total_time})
    return stage_times_summary

# Hàm điều chỉnh thời gian công việc dựa trên độ thành thạo
def adjust_task_time_with_proficiency(individual):
    worker_times = {worker: 0 for worker in workers}
    for task_idx, worker_list in enumerate(individual):
        task_time = dt['TGLT'][task_idx]
        n_workers = len(worker_list)
        theoretical_time_per_worker = task_time / n_workers
        for worker in worker_list:
            worker_proficiency = proficiency_levels[worker]
            worker_time = theoretical_time_per_worker * (1 / worker_proficiency)
            worker_times[worker] += worker_time
    return worker_times

# Phân bổ công nhân vào các công đoạn theo các ràng buộc
def select_workers_for_task(task_idx):
    machine_type = dt['THIẾT BỊ'].iloc[task_idx]
    available_workers = []

    # Lọc các công nhân khả dụng theo ràng buộc loại thiết bị
    for worker in workers:
        assigned_machines = worker_machine_assignment[worker]

        # Nếu thiết bị là loại 0, công nhân có thể làm tối đa 3 loại máy
        if machine_type == 0:
            if len(assigned_machines) < 3 or machine_type in assigned_machines:
                available_workers.append(worker)

        # Nếu thiết bị không phải là loại 0, công nhân chỉ có thể làm tối đa 2 loại máy
        else:
            if 0 in assigned_machines:
                if len(assigned_machines) < 3 or machine_type in assigned_machines:
                    available_workers.append(worker)
            else:
                if len(assigned_machines) < 2 or machine_type in assigned_machines:
                    available_workers.append(worker)

    # Đảm bảo công đoạn có ít nhất 1 công nhân
    if not available_workers:
        # Nếu không có công nhân nào khả dụng, chọn ngẫu nhiên một công nhân
        selected_workers = np.random.choice(workers, 1, replace=False)
    else:
        # Giới hạn tối đa 3 công nhân cho mỗi công đoạn
        n_selected_workers = min(len(available_workers), random.randint(1, 3))
        selected_workers = np.random.choice(available_workers, n_selected_workers, replace=False)

    # Cập nhật trạng thái phân công của các công nhân đã chọn
    for worker in selected_workers:
        worker_machine_assignment[worker].add(machine_type)
        worker_task_count[worker] += 1
        worker_assigned[worker] += 1

    return selected_workers


# Tạo quần thể kiến
def generate_population_ants():
    population = []
    for _ in range(num_ants):
        individual = []
        for task_idx in range(len(tasks_times)):
            individual.append(select_workers_for_task(task_idx))
        population.append(individual)
    return population

# Cập nhật pheromone
def update_pheromones(best_individual):
    global pheromone_matrix
    pheromone_matrix *= (1 - evaporation_rate)
    for task_idx, worker_list in enumerate(best_individual):
        if all(len(worker_machine_assignment[worker]) <= (3 if 0 in worker_machine_assignment[worker] else 2) for worker in worker_list):
            for worker in worker_list:
                worker_idx = workers.index(worker)
                pheromone_matrix[task_idx][worker_idx] += alpha / (1 + len(worker_list))

# Định nghĩa hàm `display_stage_times` để tính toán và trả về tổng thời gian của từng công đoạn
def display_stage_times(individual):
    stage_times = []

    for task_idx, worker_list in enumerate(individual):
        task_time = dt['TGLT'].iloc[task_idx]  # Lấy thời gian lý thuyết cho công đoạn
        n_workers = len(worker_list)  # Số lượng công nhân thực hiện công đoạn

        # Tính thời gian lý thuyết chia đều cho mỗi công nhân
        theoretical_time_per_worker = task_time / n_workers

        # Tính thời gian điều chỉnh cho từng công nhân dựa trên độ thành thạo
        worker_times = []
        for worker in worker_list:
            proficiency = proficiency_levels[worker]
            worker_time = theoretical_time_per_worker / proficiency  # Điều chỉnh thời gian theo độ thành thạo
            worker_times.append(worker_time)

        # Xác định thời gian tổng cho công đoạn dựa trên số công nhân
        if n_workers == 1:
            total_time_for_task = worker_times[0]
        elif n_workers == 2:
            total_time_for_task = min(worker_times) * 2
        elif n_workers == 3:
            total_time_for_task = min(worker_times) * 3

        # Lưu thời gian cho từng công đoạn
        stage_times.append({
            'Công Đoạn': dt['TÊN CÔNG ĐOẠN'].iloc[task_idx],
            'Tổng Thời Gian': total_time_for_task
        })

    return stage_times

# Tính toán lại thời gian từng công đoạn theo nhóm
# Hàm tính tổng sản phẩm dựa trên cách chia nhóm công đoạn
# Tính tổng sản phẩm dựa trên cách chia nhóm công đoạn
def calculate_total_product_via_groups(stage_times_summary, worker_time_df):
    # Tính tổng thời gian cho từng nhóm công đoạn
    total_1 = sum(stage['Tổng Thời Gian'] for stage in stage_times_summary[0:4])       # Lấy dấu
    total_2 = sum(stage['Tổng Thời Gian'] for stage in stage_times_summary[4:27])      # Thân sau
    total_3 = sum(stage['Tổng Thời Gian'] for stage in stage_times_summary[27:50])     # Thân trước
    total_4 = sum(stage['Tổng Thời Gian'] for stage in stage_times_summary[50:53])     # Lưng
    total_5 = sum(stage['Tổng Thời Gian'] for stage in stage_times_summary[53:57])     # Quần lót
    total_6 = sum(stage['Tổng Thời Gian'] for stage in stage_times_summary[57:85])     # Hoàn thành

    # Áp dụng công thức tính tổng thời gian các công đoạn
    stage_times_df = total_1 + max(max(total_2, total_3) + total_4, total_5) + total_6

    # Lấy thời gian làm việc tối đa của công nhân từ worker_time_df
    max_worker_time = worker_time_df['Tổng Thời Gian'].max()

    # Tính tổng sản phẩm
    total_product = (28800 - stage_times_df) / max_worker_time + 1
    return total_product, stage_times_df


# Vòng lặp chính của thuật toán ACO
# Vòng lặp chính của thuật toán ACO
best_product = 0
best_individual = None
best_first_task_time = 0

for iteration in range(num_iterations):
    population = generate_population_ants()
    for individual in population:
        # Sử dụng display_stage_times để tính stage_times_summary
        stage_times_summary = display_stage_times(individual)
        worker_time_df = pd.DataFrame(adjust_task_time_with_proficiency(individual).items(), columns=['Công Nhân', 'Tổng Thời Gian'])

        # Tính total_product với calculate_total_product_via_groups đã đồng nhất công thức
        total_product, first_task_time = calculate_total_product_via_groups(stage_times_summary, worker_time_df)

        if total_product > best_product:
            best_product = total_product
            best_individual = individual
            best_first_task_time = first_task_time
    if best_individual:
        update_pheromones(best_individual)
    print(f"Thế hệ {iteration + 1}: Sản lượng tốt nhất = {best_product:.4f}")


print(f"Sản lượng tốt nhất cuối cùng: {best_product}")
```

```python
# Phần 6: Hiển thị kết quả chi tiết cho mỗi công nhân

def display_task_allocation_with_time(individual):
    detailed_task_allocation = {worker: [] for worker in workers}
    for task_idx, worker_list in enumerate(individual):
        task_time = dt['TGLT'].iloc[task_idx]
        n_workers = len(worker_list)
        theoretical_time_per_worker = task_time / n_workers
        for worker in worker_list:
            proficiency = proficiency_levels[worker]
            worker_time = theoretical_time_per_worker / proficiency
            detailed_task_allocation[worker].append(f"Công đoạn {dt['TÊN CÔNG ĐOẠN'].iloc[task_idx]}: {worker_time:.2f} giây")
    return detailed_task_allocation

task_allocation_with_time = display_task_allocation_with_time(best_individual)

worker_task_summary = []
for worker, tasks in task_allocation_with_time.items():
    total_time = sum(float(task.split(": ")[1].split()[0]) for task in tasks)
    task_details = " + ".join(tasks)
    worker_task_summary.append({
        'Công Nhân': worker,
        'Chi Tiết Công Việc': task_details,
        'Tổng Thời Gian': total_time
    })

# Chuyển thành DataFrame để hiển thị
worker_task_summary_df = pd.DataFrame(worker_task_summary)

# Hiển thị kết quả chi tiết cho mỗi công nhân
worker_task_summary_df
```

```python
# Tạo bảng chi tiết thiết bị mà mỗi công nhân đã làm
def generate_worker_device_table():
    # Liệt kê tất cả các công đoạn mà công nhân tham gia và thiết bị được sử dụng
    worker_device_details = []
    for task_idx, workers_in_task in enumerate(best_individual):
        machine_type = dt['THIẾT BỊ'][task_idx]
        task_name = dt['TÊN CÔNG ĐOẠN'][task_idx]
        for worker in workers_in_task:
            worker_device_details.append({"Công Nhân": worker, "Công Đoạn": task_name, "Thiết Bị": machine_type})

    return pd.DataFrame(worker_device_details)

# Tạo bảng và hiển thị
worker_device_table_df = generate_worker_device_table()
worker_device_table_df
```

```python
# Tạo bảng chi tiết thời gian từng công đoạn của từng công nhân
def generate_worker_task_time_table():
    # Tính thời gian từng công đoạn của mỗi công nhân dựa trên độ thành thạo
    worker_task_time_details = []
    for task_idx, workers_in_task in enumerate(best_individual):
        task_time = dt['TGLT'][task_idx]  # Thời gian lý thuyết của công đoạn
        n_workers = len(workers_in_task)
        theoretical_time_per_worker = task_time / n_workers if n_workers > 0 else 0
        for worker in workers_in_task:
            worker_proficiency = proficiency_levels[worker]
            actual_worker_time = theoretical_time_per_worker * (1 / worker_proficiency)
            worker_task_time_details.append({
                "Công Nhân": worker,
                "Công Đoạn": dt['TÊN CÔNG ĐOẠN'][task_idx],
                "Thời Gian Thực Tế": actual_worker_time
            })

    return pd.DataFrame(worker_task_time_details)

# Tạo bảng và hiển thị
worker_task_time_table_df = generate_worker_task_time_table()
worker_task_time_table_df
```
