```python

# Thông số thuật toán
num_ants = 275  # Tăng số lượng kiến để có thêm các giải pháp trong mỗi thế hệ
num_iterations = 150   # Tăng số vòng lặp để cải thiện khả năng tìm kiếm
evaporation_rate = 0.2
pheromone_initial = 0.7
pheromone_matrix = np.ones((len(dt), len(workers))) * pheromone_initial
# Hàm tính fitness
def fitness_multi_with_machine_limit(individual):
    # Kiểm tra ràng buộc công đoạn phải có ít nhất một công nhân
    for task_idx, worker_list in enumerate(individual):
        if len(worker_list) == 0:
            return None  # Vi phạm ràng buộc

    # Kiểm tra ràng buộc số loại máy công nhân có thể làm
    worker_machine_types = {worker: set() for worker in workers}
    for task_idx, worker_list in enumerate(individual):
        machine_type = dt['THIẾT BỊ'][task_idx]
        for worker in worker_list:
            worker_machine_types[worker].add(machine_type)

    penalty = 0
    for worker, machine_types in worker_machine_types.items():
        if 0 in machine_types:
            if len(machine_types) > 3:
                penalty += 1
        else:
            if len(machine_types) > 2:
                penalty += 1

    if penalty > 0:
        return None  # Vi phạm ràng buộc

    # Nếu thỏa mãn, tính fitness
    return objective_function(individual, f_0, g_0, alpha)

# Hàm mục tiêu
"""

    Parameters:
    - f_x: Số lượng sản phẩm trong giải pháp hiện tại
    - g_x: Chi phí lương trong giải pháp hiện tại
    - f_0: Số lượng sản phẩm ban đầu (tham chiếu)
    - g_0: Chi phí lương ban đầu (tham chiếu)
    - alpha: Trọng số ưu tiên (0 <= alpha <= 1)

    Returns:
    - Giá trị của hàm mục tiêu H(x) cho giải pháp hiện tại
    """
def objective_function(individual, f_0, g_0, alpha):
    f_x, _ = display_stage_times(individual)  # Chỉ lấy giá trị sản lượng (total_product)
    g_x = calculate_total_salary(individual, proficiency_levels, dt)  # Thêm các tham số
    normalized_product = f_x / f_0
    normalized_cost = g_0 / g_x
    return lamda * alpha * normalized_product + (1 - alpha) * normalized_cost

# Tính tổng thời gian và sản lượng
def display_stage_times(individual):
    stage_times = []  # Danh sách lưu tổng thời gian từng công đoạn
    worker_times = {worker: 0 for worker in workers}  # Tổng thời gian làm việc của từng công nhân

    for task_idx, worker_list in enumerate(individual):
        task_time = dt['TGLT'][task_idx]  # Thời gian lý thuyết của công đoạn
        n_workers = len(worker_list)  # Số công nhân tham gia công đoạn

        if n_workers == 0:
            continue  # Bỏ qua nếu không có công nhân

        # Tính thời gian lý thuyết chia đều cho mỗi công nhân
        theoretical_time_per_worker = task_time / n_workers

        # Tính thời gian điều chỉnh theo năng lực từng công nhân
        worker_times_per_task = [
            theoretical_time_per_worker / proficiency_levels[worker]
            for worker in worker_list
        ]

        # Xử lý khi số lượng công nhân khác nhau
        if n_workers == 1:
            total_time_for_task = worker_times_per_task[0]
        elif n_workers == 2:
            total_time_for_task = min(worker_times_per_task) * 2
        elif n_workers == 3:
            total_time_for_task = min(worker_times_per_task) * 3

        # Lưu thời gian cho công đoạn
        stage_times.append(total_time_for_task)

        # Cập nhật thời gian làm việc cho từng công nhân
        for i, worker in enumerate(worker_list):
            worker_times[worker] += worker_times_per_task[i]
     # Tạo `worker_time_df` từ thông tin chi tiết các công đoạn
    worker_task_summary = []  # Tổng thời gian cho từng công nhân
    for worker, time in worker_times.items():
        worker_task_summary.append({
            'Công Nhân': worker,
            'Tổng Thời Gian': time
        })

    worker_time_df = pd.DataFrame(worker_task_summary)
    # Tính tổng thời gian cho từng nhóm công đoạn
    total_1 = sum(stage_times[0:4])       # Lấy dấu
    total_2 = sum(stage_times[4:27])      # Thân sau
    total_3 = sum(stage_times[27:50])     # Thân trước
    total_4 = sum(stage_times[50:53])     # Lưng
    total_5 = sum(stage_times[53:57])     # Quần lót
    total_6 = sum(stage_times[57:85])     # Hoàn thành


    # Áp dụng công thức tính tổng thời gian các công đoạn
    stage_times_df = total_1 + max(max(total_2, total_3) + total_4, total_5) + total_6

    # Tính thời gian làm việc tối đa của công nhân
    max_worker_time = max(worker_times.values()) if worker_times else 0

    # Tránh lỗi chia cho 0
    if max_worker_time == 0:
        return 0, stage_times_df

    # Tính tổng sản phẩm
    total_product = (28800 - stage_times_df) / max_worker_time + 1
    return total_product, stage_times_df


# Tính tổng lương
# Tính tổng chi phí lương
def calculate_total_salary(individual):
    total_salary = 0
    for task_idx, worker_list in enumerate(individual):
        task_price = dt['ĐƠN GIÁ'][task_idx]
        price_per_worker = task_price / len(worker_list)
        for worker in worker_list:
            total_salary += price_per_worker * proficiency_levels[worker]
    return total_salary
# Phân công công nhân với ma trận pheromone
def assign_workers_aco(task_idx, pheromone_matrix, worker_machine_assignment):
    machine_type = dt['THIẾT BỊ'][task_idx]
    probabilities = []

    # Tính xác suất cho từng công nhân
    for worker_idx, worker in enumerate(workers):
        machine_types = worker_machine_assignment[worker]
        if machine_type == 0 and (len(machine_types) < 3 or machine_type in machine_types):
            probabilities.append(pheromone_matrix[task_idx][worker_idx] * proficiency_levels[worker])
        elif machine_type != 0 and (len(machine_types) < 2 or machine_type in machine_types):
            probabilities.append(pheromone_matrix[task_idx][worker_idx] * proficiency_levels[worker])
        else:
            probabilities.append(0)

    probabilities = np.array(probabilities)
    non_zero_indices = np.where(probabilities > 0)[0]

    # Nếu không có công nhân khả dụng
    if len(non_zero_indices) == 0:
        fallback_indices = sorted(range(len(workers)), key=lambda idx: proficiency_levels[workers[idx]], reverse=True)[:3]
        selected_workers = [workers[idx] for idx in fallback_indices]

    else:
        # Chuẩn hóa xác suất
        probabilities[non_zero_indices] /= np.sum(probabilities[non_zero_indices])
        size = min(len(non_zero_indices), random.randint(1, 3))
        selected_indices = np.random.choice(non_zero_indices, size=size, p=probabilities[non_zero_indices], replace=False)
        selected_workers = [workers[idx] for idx in selected_indices]

    # Cập nhật trạng thái công nhân
    for worker in selected_workers:
        worker_machine_assignment[worker].add(machine_type)

    return selected_workers



# cập nhập pheromone
def update_pheromone_aco(pheromone_matrix, population, fitness_values):
    pheromone_matrix *= (1 - evaporation_rate)  # Bốc hơi pheromone

    # Tăng mạnh pheromone cho giải pháp tốt nhất
    best_individual = population[np.argmax(fitness_values)]
    best_fitness = max(fitness_values)
    for task_idx, worker_list in enumerate(best_individual):
        for worker in worker_list:
            pheromone_matrix[task_idx][workers.index(worker)] += best_fitness


# Tính tổng chi phí lương
def calculate_total_salary(individual, proficiency_levels, dt):
    total_salary = 0
    for task_idx, worker_list in enumerate(individual):
        task_price = dt['ĐƠN GIÁ'][task_idx]
        price_per_worker = task_price / len(worker_list)
        for worker in worker_list:
            total_salary += price_per_worker * proficiency_levels[worker]
    return total_salary

def generate_random_individual():
    individual = []
    worker_machine_assignment = {worker: set() for worker in workers}
    for task_idx in range(len(dt)):
        individual.append(assign_workers_aco(task_idx, pheromone_matrix, worker_machine_assignment))
    return individual

# # Vòng lặp chính của thuật toán đàn kiến
max_best_H = -float('inf')  # Giá trị H tốt nhất ban đầu
elite_solution = None  # Giải pháp tốt nhất từ các thế hệ trước

for iteration in range(num_iterations):
    population = []  # Quần thể trong thế hệ hiện tại
    fitness_values = []  # Giá trị fitness của từng cá thể trong quần thể

    # Tạo các giải pháp cho kiến
    for ant in range(num_ants):
        individual = generate_random_individual()
        fitness = fitness_multi_with_machine_limit(individual)

        if fitness is not None:  # Chỉ thêm các giải pháp hợp lệ
            population.append(individual)
            fitness_values.append(fitness)

    # Nếu không có cá thể hợp lệ, thêm một cá thể ngẫu nhiên làm fallback
    if not fitness_values:
        fallback_individual = generate_random_individual()
        fitness = fitness_multi_with_machine_limit(fallback_individual)

        if fitness is not None:
            population.append(fallback_individual)
            fitness_values.append(fitness)

    # Giữ lại giải pháp tốt nhất từ thế hệ trước (elitism)
    if elite_solution:
        population.append(elite_solution)
        fitness_values.append(max_best_H)

    # Lấy giải pháp tốt nhất trong thế hệ hiện tại
    if fitness_values:
        best_individual = population[np.argmax(fitness_values)]
        best_H = max(fitness_values)

        # Tính sản lượng và chi phí
        total_product, _ = display_stage_times(best_individual)
        total_val = calculate_total_salary(best_individual, proficiency_levels, dt)

        print(total_product)
        print(total_val)

        # Cập nhật nếu tìm được giải pháp tốt hơn
        if best_H > max_best_H:
            max_best_H = best_H
            elite_solution = best_individual

        print(f"Thế hệ {iteration + 1}: H tốt nhất = {max_best_H:.4f}")
        print(f"Thế hệ {iteration + 1}: H  = {best_H:.4f}")

    # Cập nhật pheromone dựa trên giải pháp hiện tại
    update_pheromone_aco(pheromone_matrix, population, fitness_values)

# Sau khi kết thúc các vòng lặp, lấy giải pháp tốt nhất
best_individual = elite_solution
if elite_solution:
    print(f"Giải pháp tốt nhất cuối cùng: H = {max_best_H:.4f}")
else:
    print("Không tìm được giải pháp tốt nhất.")


```

```python
# Vòng lặp chính của thuật toán đàn kiến
def run_aco():
    max_best_H = -float('inf')  # Giá trị H tốt nhất ban đầu
    elite_solution = None  # Giải pháp tốt nhất từ các thế hệ trước

    for iteration in range(num_iterations):
        population = []  # Quần thể trong thế hệ hiện tại
        fitness_values = []  # Giá trị fitness của từng cá thể trong quần thể

        # Tạo các giải pháp cho kiến
        for ant in range(num_ants):
            individual = generate_random_individual()
            fitness = fitness_multi_with_machine_limit(individual)

            if fitness is not None:  # Chỉ thêm các giải pháp hợp lệ
                population.append(individual)
                fitness_values.append(fitness)

        # Nếu không có cá thể hợp lệ, thêm một cá thể ngẫu nhiên làm fallback
        if not fitness_values:
            fallback_individual = generate_random_individual()
            fitness = fitness_multi_with_machine_limit(fallback_individual)

            if fitness is not None:
                population.append(fallback_individual)
                fitness_values.append(fitness)

        # Giữ lại giải pháp tốt nhất từ thế hệ trước (elitism)
        if elite_solution:
            population.append(elite_solution)
            fitness_values.append(max_best_H)

        # Lấy giải pháp tốt nhất trong thế hệ hiện tại
        if fitness_values:
            best_individual = population[np.argmax(fitness_values)]
            best_H = max(fitness_values)

            # Tính sản lượng và chi phí
            total_product, _ = display_stage_times(best_individual)
            total_val = calculate_total_salary(best_individual, proficiency_levels, dt)


            # Cập nhật nếu tìm được giải pháp tốt hơn
            if best_H > max_best_H:
                max_best_H = best_H
                elite_solution = best_individual

            print("-----------")
            print("sản lượng: "+str(total_product)+" / Lương: "+str(total_val))
            print("H=", max_best_H)

        # Cập nhật pheromone dựa trên giải pháp hiện tại
        update_pheromone_aco(pheromone_matrix, population, fitness_values)

    # Sau khi kết thúc các vòng lặp, lấy giải pháp tốt nhất
    best_individual = elite_solution
    if elite_solution:
        print(f"Giải pháp tốt nhất cuối cùng: H = {max_best_H:.4f}")
    else:
        print("Không tìm được giải pháp tốt nhất.")
    return best_individual,max_best_H



```

```python
# Hàm tính lương và tổng tiền lương cho tất cả công nhân
def calculate_salary(individual, proficiency_levels, dt):
    worker_salaries = {worker: 0 for worker in workers}  # Khởi tạo lương cho mỗi công nhân bằng 0

    for task_idx, worker_list in enumerate(individual):
        # Lấy đơn giá của công đoạn hiện tại
        task_price = dt['ĐƠN GIÁ'][task_idx]
        num_workers = len(worker_list)  # Số công nhân thực hiện công đoạn

        # Chia đơn giá cho số công nhân nếu nhiều người thực hiện
        price_per_worker = task_price / num_workers

        # Tính lương cho từng công nhân tham gia công đoạn
        for worker in worker_list:
            proficiency = proficiency_levels[worker]  # Lấy năng lực của công nhân
            # Lương cho công đoạn này bằng đơn giá chia cho số công nhân, sau đó nhân năng lực
            salary_for_task = price_per_worker * proficiency
            # Cộng lương cho công đoạn này vào tổng lương của công nhân
            worker_salaries[worker] += salary_for_task

    # Tạo DataFrame chứa kết quả lương của từng công nhân
    salary_df = pd.DataFrame.from_dict(worker_salaries, orient='index', columns=['Tiền Lương'])
    salary_df.reset_index(inplace=True)
    salary_df.columns = ['Công Nhân', 'Tổng Tiền Lương']

    # Tính tổng tiền lương của tất cả công nhân
    total_salary = salary_df['Tổng Tiền Lương'].sum()
    print(f"Tổng tiền lương của tất cả công nhân: {total_salary:.2f}")

    return salary_df, total_salary
# Sử dụng hàm để tính lương cho mỗi công nhân và tổng lương
salary_df, total_salary = calculate_salary(best_individual, proficiency_levels, dt)

# Hiển thị DataFrame chứa tổng tiền lương cho mỗi công nhân
print(salary_df)
print(f"Tổng tiền lương của tất cả công nhân: {total_salary:.2f}")
```

```python
def display_stage_times(individual):
    stage_times = []

    for task_idx, worker_list in enumerate(individual):
        task_time = dt['TGLT'][task_idx]
        n_workers = len(worker_list)

        # Tính thời gian lý thuyết chia đều cho mỗi công nhân
        theoretical_time_per_worker = task_time / n_workers

        # Tính thời gian điều chỉnh cho mỗi công nhân dựa trên độ thông thạo
        worker_times = []
        for worker in worker_list:
            proficiency = proficiency_levels[worker]
            worker_time = theoretical_time_per_worker / proficiency  # Điều chỉnh thời gian theo độ thông thạo
            worker_times.append(worker_time)

        # Nếu có 1 công nhân, giữ nguyên thời gian
        if n_workers == 1:
            total_time_for_task = worker_times[0]
        # Nếu có 2 công nhân, lấy người nhanh nhất và nhân đôi thời gian
        elif n_workers == 2:
            total_time_for_task = min(worker_times) * 2
        # Nếu có 3 công nhân, lấy người nhanh nhất và nhân ba thời gian
        elif n_workers == 3:
            total_time_for_task = min(worker_times) * 3

        # Lưu trữ thời gian cho từng công đoạn
        stage_times.append({
            'Công Đoạn': dt['TÊN CÔNG ĐOẠN'].iloc[task_idx],
            'Tổng Thời Gian': total_time_for_task
        })

    return stage_times

# Sử dụng hàm display_stage_times để hiển thị thời gian của từng công đoạn
stage_times_summary = display_stage_times(best_individual)

# Chuyển thành DataFrame để hiển thị
stage_times_df = pd.DataFrame(stage_times_summary)

# Hiển thị kết quả
stage_times_df
```
